# Combining Boolean Masks

```python
b = np.arange(-10,10,1)
b.shape = (4,5)
print(b)
```

```python
# find the location of the value 4
print(np.where(b==4))
print(b[2,4])
loc = np.where(b==4)
print(loc)
print(type(loc))
```

```python
mask = b==8
print(mask)
print(f'here is the value{b[mask]}')
```

```python
# print values in the last colum which are greater than 0
print(b)
mask2 = b[:,-1]>0
print(mask2)
print(b[mask2],'\n')
print(b[:,-1][mask2])
```

```python
# print out all values 5<x<-7
mask3 = b>5
print(mask3)
mask4 = b<-7
print(mask4)
mask5 = np.logical_or(mask3,mask4)
print('\n', b[mask5])
```

```python
# print out the values -4 <= x<= 2
mask6 = b>=-4
mask7 = b<=2
mask8 = np.logical_and(mask6,mask7)
print(b[mask8])
```

# Line Plot

```python
#imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# data
t = [2,4,11,13]

# figure, axes objects
fig, ax = plt.subplots()

# plot the data
ax.plot(t)

# customization

#show the plot
plt.show()
```

```python
# short version,  line plot x,y points
x = np.array([2,4,6,8,10])
y = np.array([1,3,4,7,13])
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
# customized plot
x = np.arange(0,100)
y = x*2+10
fig, ax = plt.subplots()
ax.plot(x,y,'g')               #color of line

#custom
ax.set_title('custom plot')    #title
ax.set_xlabel('x-axis')        #axes labels
ax.set_ylabel('y-label')

plt.show()
```

# line plot, two lines, legend

```python
x = np.arange(0,np.pi*4,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x,y1,'olivedrab',x,y2,'bisque')

ax.set_title('Sine and Cosine Functions')
ax.set_xlabel('0 - 4pi')
ax.set_ylabel('sine and cosine of x')
ax.legend(['sine','cosine'])

plt.savefig('plot_with_two_lines.png', dpi= 500, bbox_inches='tight')
plt.show()
```

# limit range of axes

```python
x = np.arange(0,8*np.pi,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig, ax = plt.subplots()
ax.plot(x,y1,x,y2,x,y3)

ax.legend(['sin(x)', 'cos(x)', 'tan(x)'])
ax.set_xlim([0,20])
ax.set_ylim([-4,4])

plt.savefig('plot_with_3_lines.png', dpi=500, bbox_inches='tight')
plt.show()
```
