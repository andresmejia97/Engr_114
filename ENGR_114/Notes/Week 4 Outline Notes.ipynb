```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
h = [3,11,6]
bar_locs = [0,1,2]
bar_labels = ['less than \n 20 mins', 'between 20 and \n 40 mins', 'greater than \n 40 min']

fig, ax = plt.subplots()

ax.bar(bar_locs, h, width = 0.5)  # default align = center

ax.set_title('set of commute times')
ax.set_xticks(bar_locs)     # change x axis ticks
ax.set_xticklabels(bar_labels)

plt.show()
```

# chart with ax.pie

```python
# Windows = 13
# MacOS = 4
# Chromebook = 2

sizes = [13,4,2]
labels = ['Windows', "MacOS", 'Chromebook']
explode = [0.1,0.1,0.5]  # how far out from center of pie

fig, ax = plt.subplots()
ax.pie(sizes, labels = labels, autopct = '%1.0f%%', explode = explode)

ax.axis('equal')
ax.set_title('Pie chart of computer type')

plt.show()
```

# histogram with ax.hist

```python
mu = 20
sigma = 5
x = np.random.normal(mu,sigma, size = 1000)  # 1000 random numbers weighted by previous ints

fig, ax = plt.subplots()
ax.hist(x, bins = 10)   # bins = amount of bars

ax.set_xlabel('bin range')
ax.set_ylabel('frequency')
ax.set_title('a histogram')

plt.show()
```

# scatter plot with ax.scatter

```python
n = 300
x1 = 2*np.random.randn(n) +5
x2 = 2*np.random.randn(n) +3
y1 = 4*np.random.randn(n) + 20
y2 = 4*np.random.randn(n) + 12

x = np.append(x1,x2)
y = np.append(y1,y2)

fig, ax = plt.subplots()

ax.scatter(x,y, alpha=0.2)

ax.set_title('scatter plot with two semi random arrays')

plt.show()
```

# yy plots

graphs with multiple y values that are very different

```python
# subplot
x = np.arange(0,8,0.1)
y1 = np.sin(x)
y2 = np.exp(x)

fig, ax1 = plt.subplots()

ax1.plot(x,y1)
ax1.set_xlabel('x from 0 to 8')
ax1.set_ylabel('sin(x)')
ax1.set_title('y-y plot of sin(x) and exp(x)')

ax2 = ax1.twinx()
ax2.plot(x,y2,'red')
ax2.set_ylabel('exp(x)')

plt.show()
```

# build subplot from external file

```python
x = np.arange(0,10,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# 3 axis object, one for each subplot
fig, [ax1,ax2,ax3] = plt.subplots(nrows=1,ncols=3, figsize=(12,3))

ax1.plot(x,y1)
ax1.set_title('sin(x)')

ax2.plot(x,y2)
ax2.set_title('cos(x)')

ax3.plot(x,y3)
ax3.set_title('tan(x)')

plt.show()
```
