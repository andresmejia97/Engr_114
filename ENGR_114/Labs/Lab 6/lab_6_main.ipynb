```python
image = plt.imread('stinkbug.png')
fig, ax = plt.subplots(figsize=(5,5))

color_in = input('what color map do you want?')
if color_in in ['grey','Grey','gray','Gray','greys','Greys']:
    ax.imshow(image, cmap='Greys')
    print('cmap = grey')
elif color_in in ['ocean','Ocean']:
    ax.imshow(image, cmap='ocean')
    print('cmap = ocean')
elif color_in in ['gist_earth','earth','Earth']:
    ax.imshow(image, cmap='gist_earth')
    print('cmap = gist_earth')
elif color_in in ['terrain','Terrain']:
    ax.imshow(image, cmap='terrain')
    print('cmap = terrain')
else:
    print(f'input \'{color_in}\' not valid. No cmap selected')
    ax.imshow(image)

t_labels = input('Do you want tick labels?')
if t_labels in ['no','No','nope','nah']:
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    print('tick labels = no')
elif t_labels in ['yes','Yes','y','Y']:
     ax.set_xlabel('pixels')
     ax.set_ylabel('pixels')
     print('tick labels = yes')
else:
    print(f'input \'{t_labels}\' invalid. No tick labels displayed')
    ax.set_xticklabels([])
    ax.set_yticklabels([])

grid_in = input('Do you want a grid?')
if grid_in in ['yes','Yes','y','Y']:
    ax.grid()
    print('grid = yes')
elif grid_in in ['no','No','nope','nah']:
    print('No grid selected')
else: 
    print(f'{grid_in} not valid input. No grid selected')


plt.show()
```
