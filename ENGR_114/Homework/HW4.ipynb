```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10,1)
y = np.sqrt(2*x)

fig, ax = plt.subplots()
ax.plot(y)
ax.grid()

plt.show()
```

```python
# 6.09
t = np.arange(0,20,1)
y1 = np.cos(t)
y2 = np.cos(t/2)
y3 = 1/2 * np.cos(t)

fig, ax = plt.subplots()
ax.plot(t,y1,'r')
ax.plot(t,y2,'b')
ax.plot(t,y3,'g')
ax.grid()
ax.legend(['cos(t)','cos(t/2)','1/2 cos(t)'])

plt.show()
```
